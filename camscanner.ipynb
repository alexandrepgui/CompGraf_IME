```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img_path = 'image.png'
img = cv2.imread(img_path)
A = (345, 654)
B = (1940, 416)
C = (2860, 2583)
D = (490, 3118)
input_points = np.float32([A, B, C, D])
w, h, _ = img.shape
```

```python
E = (0, 0)
F = (w, 0)
G = (w, h)
H = (0, h)
output_points = np.float32([E, F, G, H])
```

```python
M = cv2.getPerspectiveTransform(input_points,output_points)
out = cv2.warpPerspective(img,M,(w, h),flags=cv2.INTER_LINEAR)

plt.figure(figsize = (21,30))
plt.imshow(out)
```
